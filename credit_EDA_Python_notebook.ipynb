```python
#Age in days are all negative values. Min and max are negative
appl.DAYS_BIRTH.describe()
```

```python
#Lets convert it to positive and convert from days to years
appl1['AGE']=round(((appl1.DAYS_BIRTH*-1)/365),2)
appl1.AGE.describe()
```

```python
#Putting age into buckets
#defining function
def split_age(x):
    if (x>=20 and x<30):
        return '20to30'
    elif (x>=30 and x<40):
        return '30to40'
    elif (x>=40 and x<50):
        return '40to50'
    elif (x>=50 and x<60):
        return '50to60'
    else:
        return 'above60'
#new column age_bin using the function
appl1['AGE_BIN']=appl1.AGE.apply(split_age)
appl1.AGE_BIN.value_counts().plot.bar()
plt.title('Different Age buckets\n', fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Green'})
plt.show()
```

```python
#Education to target
res_ed=pd.pivot_table(data=appl1, index='NAME_EDUCATION_TYPE', columns='TARGET', values='SK_ID_CURR', aggfunc='count')
res_ed[[0, 1]].plot(kind = 'barh', stacked = True, title = 'Education to Target Analysis')
plt.legend(["Payment on Time", "Payment delayed"], loc='lower right')
plt.show()
```

```python
#Since the count is different for different category, lets normalise the data
res_ed.reset_index(inplace=True)
res_ed.set_index('NAME_EDUCATION_TYPE', inplace=True)
res_ed['Total']=res_ed[0]+res_ed[1]
res_ed['0_per']=res_ed[0]/res_ed['Total']*100
res_ed['1_per']=res_ed[1]/res_ed['Total']*100
res_ed[['0_per', '1_per']].plot(kind = 'barh', stacked = True, title = 'Education to Target Analysis')
plt.legend(["Payment on Time", "Payment delayed"], loc='upper left')
plt.show()
```

```python
#Housing type to Target
res_hous=pd.pivot_table(data=appl1, index='NAME_HOUSING_TYPE', columns='TARGET', values='SK_ID_CURR', aggfunc='count')
res_hous.reset_index(inplace=True)
res_hous.fillna(0, inplace=True)
res_hous.set_index('NAME_HOUSING_TYPE', inplace=True)
res_hous['Total']=res_hous[0]+res_hous[1]
res_hous['0_per']=res_hous[0]/res_hous['Total']*100
res_hous['1_per']=res_hous[1]/res_hous['Total']*100
res_hous[['0_per', '1_per']].plot(kind = 'barh', stacked = True, title = 'Housing Type to Target Analysis')
plt.legend(["Payment on Time", "Payment delayed"], loc='upper left')
plt.show()
```

```python
#Age to TARGET.
res_age=pd.pivot_table(data=appl1, index='AGE_BIN', columns='TARGET', values='SK_ID_CURR', aggfunc='count')
res_age.reset_index(inplace=True)
res_age.fillna(0, inplace=True)
res_age.set_index('AGE_BIN', inplace=True)
res_age['Total']=res_age[0]+res_age[1]
res_age['0_per']=res_age[0]/res_age['Total']*100
res_age['1_per']=res_age[1]/res_age['Total']*100
res_age[['0_per', '1_per']].plot(kind = 'barh', stacked = True, title = 'Age to Target Analysis')
plt.legend(["Payment on Time", "Payment delayed"], loc='upper left')
plt.show()
```
